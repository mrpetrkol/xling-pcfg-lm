```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
from matplotlib.ticker import ScalarFormatter
from matplotlib.colors import LinearSegmentedColormap


plt.rcParams['figure.dpi'] = 150

df = pd.read_csv('wandb_export_2025-04-23T16_49_33.596+02_00.csv')
df = df[~df['Name'].isin([
    "__ref__broken sandy-microwave-179",
    "amber-night-178",
    "visionary-silence-177",
    "fast-salad-176",
])]

df2 = pd.read_csv('wandb_export_2025-04-23T19_15_10.146+02_00.csv')
df2 = df2[~df2['Name'].isin([
    "__ref__broken sandy-microwave-179",
    "amber-night-178",
    "visionary-silence-177",
    "fast-salad-176",
])]


plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots()

norm = plt.Normalize(0.0, 1.0)
# norm = mcolors.TwoSlopeNorm(vmin=-0.4, vcenter=0.5, vmax=1.4)
# cmap = cm.get_cmap('coolwarm')
# colors = ['red', 'darkgray', 'blue']
# colors = ['darkblue', 'lightblue', 'darkgray', 'mistyrose', 'darkred']
colors = ['darkred', 'mistyrose', 'darkgray', 'lightblue', 'darkblue']
cmap = LinearSegmentedColormap.from_list('custom_coolwarm', colors)
cmap2 = cm.get_cmap('viridis_r')


for label, group in sorted(df.groupby('p'), reverse=True):
    group_sorted = group.sort_values('num_rules_to_swap')
    # alpha = norm(label)
    color = cmap(norm(label))
    ax.plot(group_sorted['num_rules_to_swap'], group_sorted['cross_entropy_lm_en2'], linestyle='--', marker='o', linewidth=1, color=color, label=f'(% of en2): 1 - p={round(1 - label, 2)}')

for label, group in sorted(df2.groupby('p'), reverse=True):
    group_sorted = group.sort_values('num_rules_to_swap')
    # alpha = norm(label)
    color = cmap2(norm(label))
    ax.plot(group_sorted['num_rules_to_swap'], group_sorted['cross_entropy_lm_en1'], linestyle='--', marker='o', linewidth=1, color=color, label=f'(% of en1): p={round(label, 2)}')


ax.set_xscale('log')
unique_X_values = sorted(df['num_rules_to_swap'].unique())
ax.set_xticks(unique_X_values)
ax.xaxis.set_major_formatter(ScalarFormatter())
ax.minorticks_off()

plt.xlabel('num_rules_to_swap')
plt.ylabel('cross_entropy_lm')
# plt.title('cross_entropy_lm_en2 vs num_rules_to_swap grouped by C')
plt.legend()
# plt.show()

plt.savefig('nrules_vs_ce_all.png', dpi=300, bbox_inches='tight')
```
